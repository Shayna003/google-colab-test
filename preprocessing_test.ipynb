preprocessing using nltk library

```python
nltk.download('stopwords')
nltk.download('wordnet')
lemmatizer = nltk.stem.WordNetLemmatizer()
tokeniser = nltk.tokenize.WhitespaceTokenizer()
```

```python
file_url = "https://raw.githubusercontent.com/Shayna003/google-colab-test/main/preprocessing_test.csv"
data_file = pd.read_csv(file_url) #reads dataset

```

```python
def pre_processing():
    nlp = spacy.blank('en')
    data_file['text'] = data_file['text'].astype(str).str.lower() #makes all reviews lowercase
    punct = "\n\r" + string.punctuation
    data_file['text'] = data_file['text'].str.translate(str.maketrans('', '', punct)) #removes all punctuation
    stop = stopwords.words('english')
    data_file['text'] = data_file['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)])) # removes stopwords

    data_file['text'] = data_file['text'].map(lemmatisation)  #lemmatisation : changing words with same meaning into one word

def lemmatisation(text):
    s = ''
    for w in tokeniser.tokenize(text):
      s += lemmatizer.lemmatize(w)
      s += ' '
    s = s[:-1]
    return s
```
